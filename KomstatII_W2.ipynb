## Data Visualization
Data visualization is the graphical representation of information and data. It transforms complex data sets into simpler visual formats, making them easier to understand. By using visual elements, it helps identify trends, patterns, and outliers clearly. In business, data visualization enables decision-makers to view analytics visually, supporting more accurate and efficient decision-making.

Python's libraries for data visualization:

▪ **Matplotlib** (for this week)

▪ **Seaborn** (for this week)

▪ Plotly

▪ Bokeh

▪ Altair

```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
url = "https://raw.githubusercontent.com/novrian6/dataanalysis-dataset/master/WHO-COVID-19-global-data.csv"
df = pd.read_csv(url, delimiter=';')

# Show first five rows
print(df.head())
```

## Visualization Examples with Matplotlib

```python
# Line Plot
# Convert 'Date_reported' to datetime format
df['Date_reported'] = pd.to_datetime(df['Date_reported'], format='%d/%m/%Y')
df_country = df[df['Country'] == 'Italy']
plt.figure(figsize=(12, 6))
plt.plot(df_country['Date_reported'], df_country['Cumulative_cases'], marker='o',
linestyle='-', color='blue', label='Cumulative Cases')
plt.xlabel('Date')
plt.ylabel('Cumulative Cases')
plt.title('Cumulative COVID-19 Cases in Italy')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bar plot
# Plot total new cases by country
top_countries = df.groupby('Country')['New_cases'].sum().nlargest(10)
plt.figure(figsize=(14, 8))
top_countries.plot(kind='bar', color='red')
plt.xlabel('Country')
plt.ylabel('New Cases')
plt.title('Top 10 Countries by New COVID-19 Cases')
plt.xticks(rotation=75)
plt.show()
```

```python
# Histogram
# Distribution of new cases
plt.figure(figsize=(12, 6))
plt.hist(df['New_cases'], bins=10, color='plum', edgecolor='black', alpha=0.7) # bins = x artinya nilai dibagi jadi x kelompok
plt.xlabel('New Cases')
plt.ylabel('Frequency')
plt.title('Distribution of New COVID-19 Cases')
plt.grid(True)
plt.show()
```

```python
# Scatter Plot
# Scatter plot of new cases vs. new deaths
plt.figure(figsize=(12, 6))
plt.scatter(df['New_cases'], df['New_deaths'], alpha=0.5, color='magenta')
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.title('Scatter Plot of New Cases vs New Deaths')
plt.grid(True)
plt.show()
```

## Visualization Examples with Seaborn

```python
# Line Plot with Seaborn
# Load the 'flights' dataset
df_flights = sns.load_dataset('flights')
df_flights.head()

# Line plot of year vs. passengers
sns.relplot(x='year', y='passengers', data=df_flights,
kind='line')
plt.show()
```

```python
# Bar Plot with Seaborn
plt.figure(figsize=(14, 8))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='Reds')

plt.title('Top 10 Countries by New COVID-19 Cases (Seaborn)')
plt.xlabel('Country')
plt.ylabel('New Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Histogram with Seaborn
plt.figure(figsize=(12, 6))
sns.histplot(df['New_cases'], bins=30, kde=True, color='green')
plt.title('Distribution of New COVID-19 Cases (Seaborn)')
plt.xlabel('New Cases')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot with Seaborn
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='New_cases', y='New_deaths',
alpha=0.6, color='purple')
plt.title('Scatter Plot of New Cases vs New Deaths (Seaborn)')
plt.xlabel('New Cases')
plt.ylabel('New Deaths')
plt.show()
```
